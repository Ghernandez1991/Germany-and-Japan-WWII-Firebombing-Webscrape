```python
import pandas as pd
from bs4 import BeautifulSoup
from splinter import Browser
executable_path = {"executable_path": "chromedriver.exe"}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
url = 'https://en.wikipedia.org/wiki/Strategic_bombing_during_World_War_II#U.S._bombing_of_Japan'
#call the browers and go to the url
browser.visit(url)
```

```python
tables = pd.read_html(url)
```

```python
tables[3]
```

```python
tables[7]
```

```python
germany_table = tables[3]
germany_table.columns = ['City', 'Percent_Destroyed']
```

```python
japan_table = tables[7]
japan_table.columns = ['City', 'Percent_Destroyed']
japan_table
```
